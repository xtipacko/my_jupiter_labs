```python
import matplotlib
import matplotlib.pyplot as plt
from numpy import sin,cos,pi,e,array
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
plt.clf()
fig, ax = plt.subplots(figsize=(10, 10), dpi=80)
ax.axis([-1.5,1.5,-1.5,1.5])
ax.spines['bottom'].set_position('center')
ax.spines['left'].set_position('center')
ax.spines['top'].set_color((1,1,1))
ax.spines['right'].set_color((1,1,1))


rad = lambda grad: (pi/180)*grad*15
r = array([1,0])
i = array([0,1])
plot, = ax.plot([], [], color='g', markerfacecolor='g', markeredgecolor='grey', markeredgewidth=3, markersize=30, marker='.')

def init():
    plot.set_data([],[])
    return(plot,)

def animate(t):
    x = rad(t)
    z = r*cos(x) + i*sin(x)
    #print(z)
    plot.set_data(*z)
    return (plot,)

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=25, interval=77, blit=True)
HTML(anim.to_html5_video())
```
