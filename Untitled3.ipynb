```python
import plotly
import numpy as np
from plotly.graph_objs import Scatter, Layout

plotly.offline.init_notebook_mode(connected=True)

N = 1000
random_x = np.random.randn(N)
random_y = np.random.randn(N)

trace = Scatter(
    x = random_x,
    y = random_y,
    mode = 'markers'
)

data = [trace]

plotly.offline.iplot({
    "data": data,
    "layout": Layout(title="hello world")
})
```
